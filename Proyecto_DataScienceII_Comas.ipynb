```python
print("Variables numéricas")
df_numerico= df.select_dtypes(include='number').iloc[:,1:]
df_numerico.head()
```

```python
print("matriz de correlación")
matriz_corr_df= df_numerico.corr()
matriz_corr_df.round(3)
```

```python
print("Heatmap - Matriz de correlación")
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(matriz_corr_df, annot=True, fmt=".2f",ax=ax )
plt.title("Matriz de Correlación")
plt.show()
```

```python
df[['Ingrediente_1','Ingrediente_2','Ingrediente_3']] = df['Ingredient_Ratio'].str.split(':', expand=True)
```

```python
#Convertir Brew_date a datetime
df['Brew_Date'] = pd.to_datetime(df['Brew_Date'])
```

```python
#Crear columnas de tiempo
df['Year'] = df['Brew_Date'].dt.year
df['Month'] = df['Brew_Date'].dt.month
df['Day'] = df['Brew_Date'].dt.day
```

```python
print("Distribución de variables numéricas")
df[['Fermentation_Time',
    'Temperature',
    'pH_Level',
    'Gravity',
    'Alcohol_Content',
    'Bitterness',
    'Color',
    'Volume_Produced',
    'Total_Sales',
    'Quality_Score',
    'Brewhouse_Efficiency',
    'Loss_During_Brewing',
    'Loss_During_Fermentation',
    'Loss_During_Bottling_Kegging'
    ]
   ].hist(figsize=(25, 15))
```

```python
print("Distribución de variables categóricas")
df['Beer_Style'].value_counts().plot(kind='bar')
df['Location'].value_counts().plot(kind='bar')
```

```python
print("Relaciones entre variables")
sns.scatterplot(x='Alcohol_Content', y='Quality_Score', data=df)
sns.scatterplot(x='Temperature', y='Fermentation_Time', data=df)
sns.boxplot(x='Beer_Style', y='Total_Sales', data=df)
```

```python
print("Ventas y producción")
df.groupby(['Year', 'Month'])['Total_Sales'].sum().plot()
df.groupby(['Year', 'Month'])['Volume_Produced'].sum().plot()
```

```python
ventas_estilo_ubicacion = df.groupby(['Beer_Style', 'Location'])['Total_Sales'].agg(['sum', 'mean'])
```

```python
print("Ventas por estilo y ubicación")
ventas_estilo_ubicacion = df.groupby(['Beer_Style', 'Location'])['Total_Sales'].sum()
print(ventas_estilo_ubicacion)
```

```python
print("Ventas en tiempo")
ventas_mes = df.groupby(['Year', 'Month'])['Total_Sales'].sum()
ventas_mes.plot()
```

```python
print("Ventas por SKU")
Ventas_SKU = df.groupby(['SKU'])['Total_Sales'].sum()
print(Ventas_SKU)
```

```python
print("Relación entre parámetros y calidad")
sns.scatterplot(x='Temperature', y='Quality_Score', data=df)
sns.scatterplot(x='pH_Level', y='Quality_Score', data=df)
sns.scatterplot(x='Alcohol_Content', y='Quality_Score', data=df)
```

```python
sns.boxplot(x='Loss_During_Brewing', y='Quality_Score', data=df)
sns.boxplot(x='Loss_During_Fermentation', y='Quality_Score', data=df)
sns.boxplot(x='Loss_During_Bottling_Kegging', y='Quality_Score', data=df)
```

```python
eficiencia_por_ubicacion = df.groupby('Location')['Brewhouse_Efficiency'].mean()
print(eficiencia_por_ubicacion)
```

```python
print("Pérdidas por etapas")
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Loss_During_Brewing', 'Loss_During_Fermentation', 'Loss_During_Bottling_Kegging']])
```

```python
numeric_cols = ['Fermentation_Time', 'Temperature', 'pH_Level', 'Gravity', 'Alcohol_Content',
                'Bitterness', 'Color', 'Volume_Produced', 'Total_Sales', 'Quality_Score',
                'Brewhouse_Efficiency', 'Loss_During_Brewing', 'Loss_During_Fermentation',
                'Loss_During_Bottling_Kegging', 'Ingredient_1', 'Ingredient_2', 'Ingredient_3']

df_numeric = df[numeric_cols]

correlation_matrix = df_numeric.corr()

mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

plt.figure(dpi=120, figsize=(12, 10))

sns.heatmap(correlation_matrix,
            mask=mask,
            fmt=".2f",
            annot=True,
            lw=1,
            cmap='viridis',
            center=0)

plt.yticks(rotation=0)
plt.xticks(rotation=45, ha='right')
plt.title('Matriz de Correlación entre Variables Numéricas', fontsize=16)

plt.show()
```
